# Euclidian function for calculating distance between datapoints

```python
def euclidDistance(C, X):
    x = C[1]
    x1 = X[1]
    y1 = X[2]
    y = C[2]
    ans1 = x - x1
    ans2 = y - y1
    ans1 = ans1 * ans1
    ans2 = ans2 * ans2
    ans = ans1 + ans2
    ans = math.sqrt(ans)
    return ans
```
